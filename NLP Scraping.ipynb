```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
```

```python
url = 'https://www.sunstar.com.ph/morearticles/baguio/local-news'

driver = webdriver.Chrome()
driver.get(url)
```

```python
articles = driver.find_elements(By.CLASS_NAME, 'col-md-3')
list_articles = []
for article in articles:
    title = article.find_element(By.CLASS_NAME, 'article-title').text
    link = article.find_element(By.TAG_NAME, 'a')
    date = article.find_element(By.CLASS_NAME, 'article-date').text

    newl = link.get_attribute('href')

    driver.implicitly_wait(10)
    driver.execute_script("window.open('');")
    driver.switch_to.window(driver.window_handles[1])
    driver.get(newl)
    driver.implicitly_wait(10)

    content = driver.find_element(By.CLASS_NAME, 'article-body').text

    per_article = {
        'title': title,
        'date': date,
        'content': content
    }

    list_articles.append(per_article)
    df = pd.DataFrame(list_articles)

    driver.close()
    driver.switch_to.window(driver.window_handles[0])

print(df)
```

### BAGUIO MIDLAND COURIER

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
from selenium.common.exceptions import NoSuchElementException

url = 'https://www.baguiomidlandcourier.com.ph/araw-ng-kagitingan/'

driver = webdriver.Firefox()
driver.get(url)

data_dictionary = {'Title': [], 'Date': [], 'Content': []}
```

```python
for i in range(500):
    title = ''
    date = ''
    content =''
    try:
        title = driver.find_element(By.CLASS_NAME, 'entry-title').text
        date = driver.find_element(By.CLASS_NAME, 'entry-meta').text
        try:
            content = driver.find_element(By.CLASS_NAME, 'wp-element-caption').text # For captions
        except NoSuchElementException:
            pass
        contents = driver.find_elements(By.TAG_NAME, 'p') # For p elements

        for c in contents:
            content = content + " " + c.text

    except NoSuchElementException:
        pass
    
    try:
        link = driver.find_element(By.CLASS_NAME, 'nav-previous')
        next = link.find_element(By.TAG_NAME, 'a')
        newl = next.get_attribute('href')

        data_dictionary['Title'].append(title)
        data_dictionary['Date'].append(date)
        data_dictionary['Content'].append(content)

        driver.implicitly_wait(10)
        driver.get(newl)
        driver.implicitly_wait(10)
    except NoSuchElementException:
        break
    
driver.close()
```

```python
df = pd.DataFrame(data_dictionary)
df.to_csv('data/BaguioMidlandCourier.csv')
df
```
